```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
import xgboost as xgb
from sklearn.model_selection import train_test_split
import seaborn as sns
from Handling_Outliers import Handling_Outliers
from sklearn.metrics import mean_squared_error
```

```python
df_train,df_test = pd.read_csv('train.csv'),pd.read_csv('test.csv')
```

```python
df_train.head(1)
```

```python
lb = LabelEncoder()
ss = StandardScaler()
```

```python
def to_lb(df,col):
    df[f'{col}'] = lb.fit_transform(df[f'{col}'])
    
```

```python
to_lb(df_train, 'accident')
to_lb(df_test, 'accident')
```

```python
df_id = pd.DataFrame({
    'id':df_test['id']
})
```

```python
df_train.drop(['brand', 'model', 'engine', 'ext_col', 'int_col', 'id', 'milage', 'fuel_type', 'transmission', 'clean_title'], axis=1, inplace=True)
df_test.drop(['brand', 'model', 'engine', 'ext_col', 'int_col', 'id', 'milage', 'fuel_type', 'transmission', 'clean_title'], axis=1, inplace=True)
```

```python
df_train.corr()['price']
```

```python
def to_ss(df, col):
    df[col] = ss.fit_transform(df[[col]])
```

```python
to_ss(df_train, 'model_year')
# to_ss(df_train, 'accident')
to_ss(df_train, 'price')

to_ss(df_test, 'model_year')
# to_ss(df_test, 'accident')
```

```python
Handling_Outliers(df_train, 'model_year')
Handling_Outliers(df_test, 'model_year')
```

```python
df_train.corr()['price']
```

```python
X_train, y_train = df_train.drop('price', axis=1), df_train['price']
```

```python
model = xgb.XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(df_test)
```

```python
submission = pd.DataFrame({
    'id': df_id['id'],  
    'Exited': y_pred
})
```
